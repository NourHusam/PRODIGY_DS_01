```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
c=978248
columns_to_fill = [str(year) for year in range(1960, 1990)]
# Fill NaN values with approximate numbers
for column in columns_to_fill:
    df[column] = df[column].fillna(c + 33333)
    c += 33333
```

```python
df.iloc[195]
```

```python
time_periods = {
    '1960s': '1969',
    '1970s': '1979',
    '1980s': '1989',
    '1990s': '1999',
    '2020s': '2023'
}

New_df = df[['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code']].copy()

for period_name, last_year in time_periods.items():
    New_df[period_name] = df[last_year]
```

```python
New_df.head()
```

```python
country = input("Enter a country name (Note you should enter the first letter capital): ") # ex : Jordan

Country = New_df[New_df['Country Name'] == country]
plt.figure(figsize=(10, 6))
plt.bar(time_periods.keys(), Country[time_periods.keys()].values.flatten(), color='b')
plt.title('Population in country')
plt.xlabel('periods')
plt.ylabel('Population')
plt.grid(True)
import matplotlib.ticker as mtick
plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))
plt.show()
```
